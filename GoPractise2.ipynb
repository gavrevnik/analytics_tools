#### Product Market Fit & Retention

Ключевой шаг перед масштабированием продукта - понять достиг ли он product market fit.  

**Product market fit** = ваш продукт лучше решает проблемы людей по сравнению с их прежними решениями.  
<u>Критерии достижения product market fit (один из двух) </u>:

1. Retention выходит на плато (часть привлеченных клиентов стабильно возвращается)  
Важна гранулярность (дни/недели/..)  - исходя из частотности проблемы людей которая решается.  
PS. полезно оценить сколько раз в месяц/год активные клиенты решают свою проблему  


2. Если Retention не выходит на плато но: его кривая/затухание >= чем у успешных конкурентов в отрасли.  
Есть бизнесы в которых невозможно достижение плато (дейтинговые сервисы, покупки квартиры, казуальные игры итд)

Два типа бизнесов по типу решаемой задачи: 

1. регулярные задачи (выход в почту, соц сети, покупка продуктов итд)    
Сложнее выиграть у старой привычки, так как там есть привыкание  
Т. е надо быть сильно лучше конкурентов на рынке
Однако далее проще удерживать засчет формирования постоянной привычки


2. нерегулярные задачи (покупка квартиры, автомобиля, организация отпуска итд)  
Легче выиграть так как пользователь каждый раз выбирает как бы заново.    
Сложнее удержать, так как в след. раз он опять будет выбирать решение без постоянной привычки.    
<u> Решение:</u> привлечь клиента в смежные сервисы с большей частотностью задач  
(пример - сервис по отслеживанию цен на жилье <-> сервис по покупке жилья)  
Перманентная реклама чтобы быть top of mind (например, aviasales)


Один продукт может иметь ретеншн на одном сегменте (страны / клиенты итд) и не иметь на другом.  
Самый простой способ - растить потенциальный рынок, на котором уже имеешь market fit  
Retention - это функция <u> от продукта и от пользователей </u>! Нерелевантные пользователи с хорошим  
продуктом также не образуют устойчивый product market fit

Подход к <u> измерению </u> product-market fit: https://gopractice.ru/stories/superhuman_engine/  
Опрос: "как люди будут себя чувствовать если их лишить данного продукта?"  
Если разочарованных >40%, то с большой вероятностью этот продукт имеет market fit в этой ЦА.  
Это примерная оценка для малых бизнесов, когда Retention пока отслеживать трудно  
(т. е оценка по прокси-метрике - насколько бизнес эффективно решает задачу клиента)  

Далее возможен следующий алгоритм для расширения сегмента с достижением market_fit:  
1. Что за клиенты - те, кто разочаровались бы без этого продукта (что их описывает)  
2. за что они любят продукт? что мешает "близким" к ним клиентам любить этот продукт?  

Далее роадмап улучшений строится над идеями улучшать то что любят, устранять из за чего не любят  
Близкие клиенты = "смежные" юзеры (это клиенты очень похожие на успешно активировавшихся).  
Полезная статья: https://andrewchen.com/the-adjacent-user-theory/  
Клиент для продукта может быть успешным (конверсия в использование, выручку, др активность) или нет.  
Успешные клиенты могут быть описаны фичами в некоем пространстве характеристик.  
Идея: находить близкие к успешным вектора пользователей (смежные клиенты), выстраивать к ним  
эмпатию и находить почему они испытывают проблемы с тем чтобы стать успешными.  

Как результат - это может изменить что то в продукте, чтобы вовлечь смежных юзеров.  
По мере роста продукта и расширения ЦА - портрет смежных юзеров будет меняться.  
Полезно брать тех смежных юзеров, изменение продукта для которых будет лежать в общей  
стратегии изменения продукта.


**Market fit - непрерывная функция**  
Чем сильнее market fit - тем чаще пользователь прибегает к продукту для решения своей задачи.  
При этом макс. частота прибегания ограничена частотой самой задачи. А это определяет Retention.  
К примеру дневной Retention для людей, покупающих продукты в среднем раз в два дня, будет ~50%  
Если сравнить два продукта <u> на одинаковой аудитории </u> решающих одинаковые задачи и  
Retention одного больше другого - то это означает, что его product-market fit сильнее.  

**Фичи внутри продукта - это тоже продукт**  
К примеру, stories - продукт в системе instagram, instagram - продукт в системе ios/android итд  
Каналы дистрибуции для продукта = точки входа в эту фичу из основного продукта.  
У разных фичей одного продукта по отношению к разной аудитории и к разным ее задачам  
может быть достигнут в разной степени product market fit.  

#### Product Channel Fit

Рост продукта определяет баланс из четырех компонентов:  

1. доступные каналы привлечения  
2. наличие привлекаемой целевой аудитории (ЦА)  
3. качество продукта (market fit на ЦА -> user retention)  
4. модель монетизации  

Лучше продукт -> больше retention  
Агрессивнее монетизация при норм. retention -> больше LTV пользователя  
Также агрессивная монетизация -> снижение ЦА (порог входа в продукт выше)  
Больше LTV -> можем привлекать более дорогих пользователей (LTV > CAC)  

Product channel fit - условие для найденного канала привлечения, когда  
1. в нем существует достаточная ЦА (зависит от продукта и порога монетизации)  
2. LTV клиента в продукте позволяет выгодно и стабильно привлекать клиентов из этой ЦА  

Сила Product market fit измеряется через качество Retention кривой клиентов в продукте.  
Сила Product channel fit - в объеме новых юзеров экономически выгодно привлекаемых из этого канала.  

На силу канала привлечения может влиять много факторов например:
1. Сам продукт и модель монетизации -> LTV -> доступные косты привлечения
2. Активность конкурентов -> разогрев канала -> повышение CAC  
3. Политика работы самих каналов (например Google Ads, Apple Aps ...)  
4. Смена паттернов самих клиентов, истощение каналов (ЦА газет постепенно снижается)  

Аналогично, теория product channel fit работает с фичами в продукте.  
Добавляя новые точки входа в продукт - можно привлекать разную ЦА (Например, stories в инстаграм)

#### Типы бизнесов и каналов масштабирования

**ЦА = B2C**
1. простой и быстрый онбординг (клиент быстро получает ценность продукта)  
2. мягкая монетизация (реклама, покупки внутри продукта, комиссии...)  
3. каналы привлечения: реферальные (виральность), органика, платная реклама ...  

**ЦА = SMB**
1. средний онбординг (нужна интеграция софта в компании)  
2. подписочная модель (~1-10k долларов месяц)
3. комбинация некоторых каналов как в B2C + команда продаж (CAC клиента выше) 
PS масс-каналы могут работать плохо, так как ЦА специфичная

**ЦА = Enterprise**  
1. сложный онбординг, пилотные проекты по внедрению  
2. дорогие контракты в качестве монетизации  
3. прямы продажи на уровне топ-менеджмента  

**Death zone**  
ИП и микро-бизнес - аудитория достаточно узкая для масс-каналов,  
клиенты не готовы платить много для того чтобы покрывать команду продаж.  
В этой зоне создается мало окупаемых бизнесов.  

<u> Типы каналов привлечения </u>  
Масштабируемые - способные дать сходимый по unit-экономики прогнозируемый рост.  
Немасштабируемые (например, пресс релиз, акция ...) - дают временную искру для развития.  

Основные масштабируемые каналы для online бизнесов:
1. реферальные программы  и пр. каналы с опорой на виральность  
2. рост органического трафика в поисковиках (SEO, ASO...)  
3. платные каналы в Facebook, Google  
4. каналы роста через команду продаж  
5. affilate marketing (взаимные реферальные программы с др. бизнесами)  


<u> Почему надо заниматься масштабированием даже если продукт уникально хороший </u>  
Изначально он будет расти засчет виральности, однако позже возникнут трудности:
1. конкуренты сделают схожее и заберут ЦА используя эффективные платные каналы продвижения  
2. конкуренты также сделают решение менее уникальным, что снизит виральность  
3. первичный гипер-рост будет засчет early adapters, далее ЦА будет менее лояльной и более требовательной  

#### Unit-экономика

Предположим, что константные расходы на продукцию равны 0 (часто в online бизнесах)  
Эффективность канала привлечения может показать ROI (return of investments)  
ROI = (GrossProfit - CAC) / CAC (во сколько раз окупаются затраты на маркетинг)  
GrossProfit = LTV -  Cost of Goods Sold  
COGS = переменные затраты на одну продажу, которые растут при масштабировании  
Например: команда поддержки, склады хранения и пр.  
PS. Иногда в online бизнесах COGS = 0 -> ROI = (LTV-CAC)/CAC
