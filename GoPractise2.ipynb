#### Модель роста

Ключевая метрика - метрика, поведение которой определяет таргет компании/команды итд. Например, выручка или MAU.  
Модель роста ключевой метрики - разложение ее на систему драйверов (более низкоуровневых метрик),  
изменение которых будет отражать изменение ключевой.  
Система драйверов может быть различной для одной и той же метрики (как разные базисы разложения вектора) -  
при этом нужно выбрать такую, на которую можно повлиять, она связана с текущими процессами продукта,  
легко интерпретируется бизнесово.
Полезная рекомендация при поиске разложения: "делайте настолько просто, насколько это возможно, но не проще".  

Один из примеров разложения (в некоторых кейсах) - система Growth Accounting.  
Аддитивная метрика (например, выручка) может быть разложена на выручку, приносимую:  
1. новыми клиентами
2. старыми возвращенными клиентами
3. ушедшими клиентами со знаком минус (используетс предикт по ср. ARPU)
