#### Модель роста - todo - это и ниже куда то в раздел рост продукта

Ключевая метрика - метрика, поведение которой определяет таргет компании/команды итд. Например, выручка или MAU.  
Модель роста ключевой метрики - разложение ее на систему драйверов (более низкоуровневых метрик),  
изменение которых будет отражать изменение ключевой.  
Система драйверов может быть различной для одной и той же метрики (как разные базисы разложения вектора) -  
при этом нужно выбрать такую, на которую можно повлиять, она связана с текущими процессами продукта,  
легко интерпретируется бизнесово.
Полезная рекомендация при поиске разложения: "делайте настолько просто, насколько это возможно, но не проще".  

Один из примеров разложения (в некоторых кейсах) - система Growth Accounting.  
Аддитивная метрика (например, выручка) может быть разложена на выручку, приносимую:  
1. новыми клиентами
2. старыми возвращенными клиентами
3. ушедшими клиентами со знаком минус (используетс предикт по ср. ARPU)

#### Добавочная ценность продукта

Как писал Морейнис - "любое новое просто более эффективно удовлетворяет старый спрос, который  
сильно не меняется во времени".  Таким образом, рост продукта - это перераспределение "работы"  
от старого продукта к новому, более эффективному.  
Пример: пленочный фотоаппарат -> цифровая камера -> смартфон  

Перераспределение "работы" от одного продукта к другому возможно только, когда новый продукт  
создает достаточную добавочную ценность относительно старого продукта, чтобы преодолеть инерцию пользователей.  
1. Яндекс Поиск не имел product market fit в Турции, так как люди там уже пользовались сравнимым по  
качеству поиска Гуглом.  
2. Яндекс Навигатор наоборот достиг свой product market fit и захватил долю рынка, так как  
карты гугла в Турции не отображали пробки и были не слишком точными.  

Product Fit достигается, когда стоимость переключения со старого продукта на новый  
ниже, чем добавочная ценность создаваемая новым решением. Отсюда путь к росту product-market fit:
1. повышать добавочную ценность (лучше продукт)
2. снижать стоимость переключения (удобства входа в новый продукт) 

Очень часто менеджеры и предприниматели при принятии решений о запуске и управлении продуктами мыслят в терминах  
динамики рынка: «Этот сегмент рынка быстро растет, нам нужно туда выйти»  
Но это очень неэффективная модель мышления, которая ведет к плохим решениям: они принимаются на основе корреляции между переменными, а не на основе причинно-следственных связей. Для того, чтобы понять, почему что-то растет, вы должны ответить на базовый вопрос: для каких пользователей и как продукт создает добавочную ценность?

<u> Хороший пример </u>  
В Азии wechat широко масштабировал технологию оплаты через мессенджер (wechat Pay).  
Повторяя за этим, Facebook пытался создать версию своей оплаты через мессенджер, но провалился.   
Дело в том что в Азии wechat создавал огромную добавочную ценность по сравнению с наличными, тогда как  
Facebook конкурировал с уже сильно развитыми на Западе VISA, Google/Apple Pay итд.  

---
Захват рынка.  
Как только продукт достиг market-fit ему необходимо максимально быстро донести это ценность до пользователей,  
чтобы последующим конкурентам (которые как минимум могут копировать решение) пришлось соревноваться с этим  
продуктом, а не со старым способом решения задачи.  
Пример Teams VS Slack, когда Microsoft уступая в продукте Слаку тем не менее обогнал его по росту, так  
как имел доступ через свою экосистему к куче компаний, которые еще не знали про Слак. На этой ЦА  
Teams соревновался со старыми способами решения задачи коммуникаций внутри компаний - письма/skype  
и легко их превзошел.  
Таким образом, Microsoft пересадив своих клиентов на более удобный продукт по сути защитил  
свою аудиторию от возможной конкуренции Slack.
