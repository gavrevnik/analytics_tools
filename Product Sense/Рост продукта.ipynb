#### Product Market Fit & Retention

Ключевой шаг перед масштабированием продукта - понять достиг ли он product market fit.  

**Product market fit** = ваш продукт лучше решает проблемы людей по сравнению с их прежними решениями.  
<u>Критерии достижения product market fit (один из двух) </u>:

1. Retention выходит на плато (часть привлеченных клиентов стабильно возвращается)  
Важна гранулярность (дни/недели/..)  - исходя из частотности проблемы людей которая решается.  
PS. полезно оценить сколько раз в месяц/год активные клиенты решают свою проблему  


2. Если Retention не выходит на плато но: его кривая/затухание >= чем у успешных конкурентов в отрасли.  
Есть бизнесы в которых невозможно достижение плато (дейтинговые сервисы, покупки квартиры, казуальные игры итд)

Два типа бизнесов по типу решаемой задачи: 

1. регулярные задачи (выход в почту, соц сети, покупка продуктов итд)    
Сложнее выиграть у старой привычки, так как там есть привыкание  
Т. е надо быть сильно лучше конкурентов на рынке
Однако далее проще удерживать засчет формирования постоянной привычки


2. нерегулярные задачи (покупка квартиры, автомобиля, организация отпуска итд)  
Легче выиграть так как пользователь каждый раз выбирает как бы заново.    
Сложнее удержать, так как в след. раз он опять будет выбирать решение без постоянной привычки.    
<u> Решение:</u> привлечь клиента в смежные сервисы с большей частотностью задач  
(пример - сервис по отслеживанию цен на жилье <-> сервис по покупке жилья)  
Перманентная реклама чтобы быть top of mind (например, aviasales)


Один продукт может иметь ретеншн на одном сегменте (страны / клиенты итд) и не иметь на другом.  
Самый простой способ - растить потенциальный рынок, на котором уже имеешь market fit  
Retention - это функция <u> от продукта и от пользователей </u>! Нерелевантные пользователи с хорошим  
продуктом также не образуют устойчивый product market fit

Подход к <u> измерению </u> product-market fit: https://gopractice.ru/stories/superhuman_engine/  
Опрос: "как люди будут себя чувствовать если их лишить данного продукта?"  
Если разочарованных >40%, то с большой вероятностью этот продукт имеет market fit в этой ЦА.  
Это примерная оценка для малых бизнесов, когда Retention пока отслеживать трудно  
(т. е оценка по прокси-метрике - насколько бизнес эффективно решает задачу клиента)  

Далее возможен следующий алгоритм для расширения сегмента с достижением market_fit:  
1. Что за клиенты - те, кто разочаровались бы без этого продукта (что их описывает)  
2. за что они любят продукт? что мешает "близким" к ним клиентам любить этот продукт?  

Далее роадмап улучшений строится над идеями улучшать то что любят, устранять из за чего не любят  
Близкие клиенты = "смежные" юзеры (это клиенты очень похожие на успешно активировавшихся).  
Полезная статья: https://andrewchen.com/the-adjacent-user-theory/  
Клиент для продукта может быть успешным (конверсия в использование, выручку, др активность) или нет.  
Успешные клиенты могут быть описаны фичами в некоем пространстве характеристик.  
Идея: находить близкие к успешным вектора пользователей (смежные клиенты), выстраивать к ним  
эмпатию и находить почему они испытывают проблемы с тем чтобы стать успешными.  

Как результат - это может изменить что то в продукте, чтобы вовлечь смежных юзеров.  
По мере роста продукта и расширения ЦА - портрет смежных юзеров будет меняться.  
Полезно брать тех смежных юзеров, изменение продукта для которых будет лежать в общей  
стратегии изменения продукта.


**Market fit - непрерывная функция**  
Чем сильнее market fit - тем чаще пользователь прибегает к продукту для решения своей задачи.  
При этом макс. частота прибегания ограничена частотой самой задачи. А это определяет Retention.  
К примеру дневной Retention для людей, покупающих продукты в среднем раз в два дня, будет ~50%  
Если сравнить два продукта <u> на одинаковой аудитории </u> решающих одинаковые задачи и  
Retention одного больше другого - то это означает, что его product-market fit сильнее.  

**Фичи внутри продукта - это тоже продукт**  
К примеру, stories - продукт в системе instagram, instagram - продукт в системе ios/android итд  
Каналы дистрибуции для продукта = точки входа в эту фичу из основного продукта.  
У разных фичей одного продукта по отношению к разной аудитории и к разным ее задачам  
может быть достигнут в разной степени product market fit.  

#### Product channel fit

Рост продукта определяет баланс из четырех компонентов:  

1. доступные каналы привлечения  
2. наличие привлекаемой целевой аудитории (ЦА)  
3. качество продукта (market fit на ЦА -> user retention)  
4. модель монетизации  

Лучше продукт -> больше retention  
Агрессивнее монетизация при норм. retention -> больше LTV пользователя  
Также агрессивная монетизация -> снижение ЦА (порог входа в продукт выше)  
Больше LTV -> можем привлекать более дорогих пользователей (LTV > CAC)  

Product channel fit - условие для найденного канала привлечения, когда  
1. в нем существует достаточная ЦА (зависит от продукта и порога монетизации)  
2. LTV клиента в продукте позволяет выгодно и стабильно привлекать клиентов из этой ЦА  

Сила Product market fit измеряется через качество Retention кривой клиентов в продукте.  
Сила Product channel fit - в объеме новых юзеров экономически выгодно привлекаемых из этого канала.  

На силу канала привлечения может влиять много факторов например:
1. Сам продукт и модель монетизации -> LTV -> доступные косты привлечения
2. Активность конкурентов -> разогрев канала -> повышение CAC  
3. Политика работы самих каналов (например Google Ads, Apple Aps ...)  
4. Смена паттернов самих клиентов, истощение каналов (ЦА газет постепенно снижается)  

Аналогично, теория product channel fit работает с фичами в продукте.  
Добавляя новые точки входа в продукт - можно привлекать разную ЦА (Например, stories в инстаграм)


Оценивая product channel fit важно учитывать такие побочные явления как:  
1. инкрементальность канала (100% если клиент не пришел бы без этой рекламы)
2. виральность продукта (повышает кол-во привлеченных за ту же цену, снижает косты)
3. корректная аттрибуция трафика (понимание что пользователь пришел именно из этого канала)


#### Типы бизнесов

**ЦА = B2C**
1. простой и быстрый онбординг (клиент быстро получает ценность продукта)  
2. мягкая монетизация (реклама, покупки внутри продукта, комиссии...)  
3. каналы привлечения: реферальные (виральность), органика, платная реклама ...  

**ЦА = SMB**
1. средний онбординг (нужна интеграция софта в компании)  
2. подписочная модель (~1-10k долларов месяц)
3. комбинация некоторых каналов как в B2C + команда продаж (CAC клиента выше) 
PS масс-каналы могут работать плохо, так как ЦА специфичная

**ЦА = Enterprise**  
1. сложный онбординг, пилотные проекты по внедрению  
2. дорогие контракты в качестве монетизации  
3. прямы продажи на уровне топ-менеджмента  

**Death zone**  
ИП и микро-бизнес - аудитория достаточно узкая для масс-каналов,  
клиенты не готовы платить много для того чтобы покрывать команду продаж.  
В этой зоне создается мало окупаемых бизнесов.  

<u> Типы каналов привлечения </u>  
Масштабируемые - способные дать сходимый по unit-экономики прогнозируемый рост.  
Немасштабируемые (например, пресс релиз, акция ...) - дают временную искру для развития.  

Основные масштабируемые каналы для online бизнесов:
1. реферальные программы  и пр. каналы с опорой на виральность  
2. рост органического трафика в поисковиках (SEO, ASO...)  
3. платные каналы в Facebook, Google  
4. каналы роста через команду продаж  
5. affilate marketing (взаимные реферальные программы с др. бизнесами)  


<u> Почему надо заниматься масштабированием даже если продукт уникально хороший </u>  
Изначально он будет расти засчет виральности, однако позже возникнут трудности:
1. конкуренты сделают схожее и заберут ЦА используя эффективные платные каналы продвижения  
2. конкуренты также сделают решение менее уникальным, что снизит виральность  
3. первичный гипер-рост будет засчет early adapters, далее ЦА будет менее лояльной и более требовательной

изменениям в продукте или в рынке, чтобы в результате открылись масштабируемые каналы

#### Unit экономика

Предположим, что константные расходы на продукцию равны 0 (часто в online бизнесах)  
Эффективность канала привлечения может показать ROI (return of investments)  
ROI = (GrossProfit - CAC) / CAC (во сколько раз окупаются затраты на маркетинг)  
GrossProfit = LTV -  Cost of Goods Sold  
COGS = переменные затраты на одну продажу, которые растут при масштабировании  
Например: команда поддержки, склады хранения и пр.  
PS. Иногда в online бизнесах COGS = 0 -> ROI = (LTV-CAC)/CAC  

<u> Масштабирование каналов с отрицательной экономикой </u>

Несколько причин, по которым компании масштабируются "в убыток":

- захват рынка с последующей монетизацией наработанной базы
- монетизация начинает работать только с крупных масштабов
также для запуска виральных механизмов и пр.
- рост капитализации (когда идет рост рынка, рост акций и деньги привлекаемые инвесторами создают нужную прибыль)

Некоторые инвестируют в каналы с ROI<0,  
учитывая поправку на виральность за счет привлекаемых пользователей.  
Однако, это нужно делать осторожно.

#### Глубина рынка

Три ключевых условия для масштабирования продукта на целевом рынке:
1. достигнут product market fit на этом рынке (retention)  
2. достигнут product channel fit (unit экономика в привлечение клиентов сходится)  
3. рынок на котором достигнуты условия 1 и 2 достаточно глубокий (большая ЦА)  

При этом выделяют **подтвержденный** рынок (рынок клиентов, на котором точно достигаются условия 1 и 2) и  
**потенциальный** рынок (это целевой, а также потенциальные смежные рынки).  
Смежными могут быть рынки других стран, схожих задач, клиентов итд, однако предстоит еще  
доказать, что на этих рынках также достигаются 1 и 2.  

На размер рынка могут влиять не только объем ЦА, но также конкуренты на этом рынке:
1. прямые (решающие тем же методом задачу клиента)
2. косвенные (решающие альтернативным)  

Обычно приток клиентов идет от косвенных конкурентов, так как с прямыми сложно соперничать,  
ибо редко получается сделать продукт многократно лучше существующего.  

Размер целевого рынка можно определять из аналитических отчетов исходя из определения  
той задачи, которую подтвердженно и с хорошим retention решаем продукт.  

(!) Расширение рынка возможно путем подключения новой ЦА, для которой мы также начинаем  
эффективно решать схожую задачу. Поэтому размер рынка правильно оценивать не исходя из простой  
категории (например - рынок такси), но из глоабьлной задачи, решаемой людьми (перемещение из А в В).  

#### МОДЕЛЬ РОСТА

Выбираем NSM метрику (подробнее см раздел "Метрики") по следующим условиям:
1. Ее рост - опережающий индикатор роста ценности бизнеса
2. Ее можно легко измерить и интерпретировать
3. Сложно "накрутить", так чтобы рост перестал быть коррелирован с ростом ценности

Определяем способность бизнеса к росту:
1. Достигнут product-market fit = плато по Retention в метрику или скорость затухания на уровне отрасли
2. Достигнут product-channel fit = есть каналы, по которым стабильный приток клиентов с ROI>0
3. Имеется достаточная глубина целевого рынка

Изучаем основные задачи, которые клиенты решают в бизнесе (см. JTBD в разделе "Customer research")   
Получаем список основных задач, которые клиенты решают, используя данный бизнес:
1. Задачи для которых клиент использует бизнес в конкретных обстоятельствах
2. Альтернативные пути которыми клиент пользуется/пользовался
3. Добавочная ценность продукта относительно альтернатив

PS Важно что задачи клиетов фундаментально не меняются со временем (передвижение, еда, общение итд)  
Меняется только контекст и постепенно улучшаются способы решения задач.  
Критерий успеха: наша добавочная ценность в решение задачи выше стоимости переключения со старого решения.

<u> Пример 1 </u>   
В Азии wechat широко масштабировал технологию оплаты через мессенджер (wechat Pay).  
Повторяя за этим, Facebook пытался создать версию своей оплаты через мессенджер, но провалился.   
Дело в том что в Азии wechat создавал огромную добавочную ценность по сравнению с наличными, тогда как  
Facebook конкурировал с уже сильно развитыми на Западе VISA, Google/Apple Pay итд.  

<u> Пример 2: захват рынка </u>  
Как только продукт достиг market-fit ему необходимо максимально быстро донести это ценность до пользователей,  
чтобы последующим конкурентам (которые как минимум могут копировать решение) пришлось соревноваться с этим  
продуктом, а не со старым способом решения задачи.  
Пример Teams VS Slack, когда Microsoft уступая в продукте Слаку тем не менее обогнал его по росту, так  
как имел доступ через свою экосистему к куче компаний, которые еще не знали про Слак. На этой ЦА  
Teams соревновался со старыми способами решения задачи коммуникаций внутри компаний - письма/skype  
и легко их превзошел.  
Таким образом, Microsoft пересадив своих клиентов на более удобный продукт по сути защитил  
свою аудиторию от возможной конкуренции Slack.

Исходя из структуры задач, составляем <u>базис</u> = систему метрик, на которую раскладываем NSM  
Условие: на метрики базиса (драйверы) можно влиять конкретными действиями команды  
Принцип разложения: "делайте настолько просто, насколько это возможно, но не проще".  

Модель роста способна показать в чем причины замедления или ускорения роста NSM,  
а также спрогнозировать динамику роста.

1. Product-fit (=качество продукта) определяет LTV пользователей (монетизация)
2. рост LTV пользователей позволяет растить макс. цену привлечения клиентов (CAC)
3. Чем выше цена привлечения - тем больше доступно клиентов в платных каналах (аукционная модель)
4. Чем больше клиентов -> тем быстрее рост базы и монетизации в будущем

<u>Окупаемость инвестиций</u>  
На привлечение клиентов тратятся деньги (чтобы покрыть CAC + переменные издержки компании)  
LTV(month_cnt) >= CAC -> month_cnt - период окупаемости привлекаемых когорт  
Пока когорта не окупилась -> привлечение убыточное и требует раунд инвестиций.  
Спустя month_cnt старые когорты бизнеса начинают приносить прибыль -> инвестиции окупаются.  

Больше окно окупаемости -> больше готовы платить за юзера (CAC) -> быстрее захват базы продуктом.  
Т.е больше инвестиции; больше срок окупаемости -> шире захват рынка (будущая защита от конкурентов)

<u> Модель роста для маркетплейсов </u>  
Структура маркетплейса обычно: 
1. сторона спроса (например, пассажиры такси) = consumers, они же платящие
2. сторона предложения (водители) = producers, редко платят
3. платформа (агрегатор) = следит за релевантностью, растет на комиссии

Ключевая NSM = GMV = суммарный объем всех сделок между агентами (все заказы в такси в месяц)  
Так как платит только сторона спроса, то LTV_demand расходуется на маркетинг и спроса и предложения.  
Юнит-экономика каналов привлечения:
1. CAC_supply_max = LTV_demand * K
2. CAC_demand_max = LTV_demand * (1-K)

K - баланс сторон спроса и предложения в маркетплейсе  
K = retention_supply / retention_demand  
Пояснение: надо привлекать клиентов с обеих сторон так, чтобы в долгую сохранялся баланс  
между стороной спроса и предложения (иначе, будет слишком много водителей и мало пассажиров и наоборот)

<u> Направления роста </u>  
Основные направления роста продуктов:  
1. На доказанном product-market-fit.  
Наиболее стабильный и предсказуемый рост, но с возможно ограниченной ЦА. Основные направления:  
улучшение эффективности продукта (рост добавочной ценности), а также работа над донесением этой  
ценности в продукте до пользователя (дизайн)  
2. На смежных рынках
Необходимо заново доказывать product/channel-fit, но потенциальная ЦА выше + есть экспертиза в домене.  
Также возможны следующие типы:  
1) по ЦА (ищем новых смежных клиентов, подстраиваем под них продукт)    
2) вертикальное (например, Apple стало само производить процессоры для смартфонов)   
3) горизонтальное (Amazon продавали книги, далее расширили линейку товаров)  
3. Через запуск новых продуктов  
Принципиально новый продукт, но разделяющий миссию компании, а также использующий накопленную экспертизу и базу.  
Пример: Яндекс.Еда, Яндекс.Такси итд  

По приоритету обычно рассматривают 1 -> 3 при условии, что на каждом уровне имеется потенциал роста.

---
<u> Стратегия роста продукта </u>   
Стратегия — это ответ, как именно компания будет достигать поставленные перед ней **цели**.  
Стратегия базируется на основе глубокого анализа и понимания бизнеса, продукта и рынка.  
Она указывает на способ, как и за счет чего компания может попасть из текущего положения в желаемое.  
Цели - значения ключевых метрик компании (удобно выражаются через NSM).  

Элементы стратегии:  
1. understand: проблемы, которые мешают прийти к цели
2. identify: общий концепт, как проблемы будут решаться
3. execute: как распределяются ресурсы, приоритеты, кто и как реализует концепт.

#### Каналы роста

В другой классификации по онлайн-каналам роста см здесь раздел "Типы бизнесов"  

<u> Линейные каналы </u>  

Каналы, в которых привлекаемый трафик линейно зависит от инвестиций в него
1. покупка по аукционной модели. Инвестиции = N клиентов * ставка за клиента
2. PR активность (статьи, блоги) - заплатили X, получили M клиентов

---
<u> Каналы с циклами роста </u>

Способ привлечения клиентов (активность), когда достигаются условия:
1. Активность привлекает когорту клиентов
2. Клиенты предсказуемо генерируют аналогичную активность
3. Данная активность вновь привлекает клиентов ...

Частный пример цикла роста: виральный цикл через сарафанное радио  
Клиент приходит в продукт, пользуется - рассказывает другим - приходят другие ...

K - коэффициент роста канала = сколько пользователей приведет один пользователь в среднем.  
Если на входе было X пользователей, то рост будет $X + X*K + (X*K)*K + ... = X(1+K+K^2+...)$  
При K > 1 рост получается экспоненциальным. При K < 1 - сумма убывающей геометрической прогрессии

Длина цикла роста = время за которое X пользователей приведет K * X других пользователей.
Чем короче длина - тем больше циклов успеет пройти за фикс. промежуток (например, год)  
Рост продукта **значительно** больше зависит от длины цикла чем от K!  

Пример: цикл роста = приходят клиенты, платят, деньги реинвестируются в их привлечение снова  
Здесь платный канал - способ которым одни клиенты приводят других.  
Если клиенты будут платить быстрее -> короче цикл -> выше скорость роста клиентов.

---
<u> Типы циклов роста </u>  

1) Виральные каналы  
Клиент приходит в продукт - затем приглашает других клиентов итд.  
1.1 Сарафанное радио - юзер верит что это полезно другому юзеру  
1.2 Органическая - юзер приглашает, чтобы увеличить ценность себе (например, в мессенджер)  
1.3 Искусственная - клиента мотивируют приглашать промокодами и пр.  
1.4 Побочная - клиенты приглашают "неумышленно" (курьеры в одежде доставки)  

2) Контентные циклы  
Юзеры создают контент или способствуют созданию - контент привлекает новых пользователей итд.  
UGC = user generated content (например, соц сети или wiki)  
CGC = company generated content (на прибыль от пользователей)  
Контентные циклы похожи на виральные, но работают вокруг контента, а не самого продукта.  

3) Платные каналы  
Закупается трафик, привлекаются пользователи, приносят прибыль, реинвестирование в трафик.

Необходимо искать в своем продукте циклы роста (обычно их зачатки уже существуют) и далее подпитывать их.  

#### Защита роста от конкурентов

Ключевые риски для роста бизнеса от конкуренции могут возникать на трех уровнях:  
1. Уровень ценности - конкуренты находят лучшее решение задачи (новое или улучшают старое)
2. Уровень каналов - каналы истощаются из за конкуренции, экономика проседает
3. Уровень рынка - конкуренты захватывают рынок, ограничивают доступную ЦА, меняют паттерны клиентов

---
<u> Механизмы защиты </u>  
1. Масштаб - когда конкуренту надо очень много проинвестировать, чтобы обогнать по уровню ценности  
(например, сложно тягаться с Amazon с его системой логистики, сложным софтом и пр.)
2. Захват рынка - бизнес донес ценность до ЦА первым и теперь чтобы с него переключиться, нужно  
преодолеть инерцию клиентов, что увеличивает стоимость их переключения на альтернативы
3. Глубокое встраивание - модель бизнеса предполагает встраивание в жизнь клиента. Например,  
базы знаний, CRM и.т.д Опять же увеличиваем стоимость переключения на альтернативы
4. Сетевые эффекты - ценность продукта для клиента растет пропорционально кол-ву участников.  
4.1 Прямой эффект - например, соц сети, где чем больше людей - тем выше ценность каждому.  
4.2 Перекрестный эффект - чем больше участников 2-ой стороны - тем лучше (продавцов на маркетплейсе).  
4.3 Эффект данных - ценность из за накопленных от участников данных (отзывы на сайте бронирования  
5. Уникальные технологии, патенты и пр.
6. Бренд - закрепление в голове клиента, также повышается стоимость переключения.

Наиболее сильные и долгосрочные механизмы защиты - сетевые, на встраивании, на масштабе, а также бренд.
