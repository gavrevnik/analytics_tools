Ключевая задача продукта - успешно решать определенную задачу (или задачи) стоящие перед клиентом.  
Важно уметь понимать, какие конкретно задачи пользователь решает с помощью продукта.  
Иногда они могут сильно отличаться от номинальных задач для которых продукт существует.  
Без правильного понимания для чего продукт используется клиентами, будет казаться, что  
широкий круг гипотез потенциально способен драйвить ключевые метрики - но часто это не так.

По мере развития компании - она часто отдаляется от пользователя и начинает оптимизировать   
абстрактный продукт, что приводит к ухудшению product-market fit для пользователей.  

Если мы хотим улучшить метрики, то мы должны фокусироваться не на самих метриках,  
а на понимании цели и сути описываемого ими процесса, и работать над его улучшением

Ключевой подход в понимании продукта - граммотный **customer research**

#### Jobs to be done

https://hbr.org/2016/09/know-your-customers-jobs-to-be-done
The “Jobs to be Done” Theory of Innovation (book)

Основная идея - сами клиенты по своему профилю могут сильно отличаться, однако при этом "нанимать"    
продукт для одной и той же своей задачи. Сегментация не по типам клиентов, а по типу решаемой задачи.  

Как только определены ключевые задачи - выясняется, какие блокеры существуют у клиентов,  
мешающие решать задачи с помощью данного продукта. Продукт модифицируется так, чтобы этих блокеров
стало меньше.

Классический пример с молочными коктейлями. Две распространенные задачи, на которые люди их "нанимают":  
1. купить детям, как средство вознаграждения за хорошее поведение и.т.д
2. купить по пути на работу в автомобиль для того чтобы быстро подкрепиться за рулем  

Для первой задачи блокеры на ее решение:  
1.1 высокое содержание сахара в коктейле (потенциальный вред)  
1.2 слишком большой размер коктейля    
1.3 альтернативные варианты поощрения (поход в магазин игрушек итд)   

Для второй задачи блокеры на ее решение:  
2.1 малая сытность коктейля  
2.2 долгое время приготовления  

Продуктовые решения для первой задачи:  
1.1 Сделать более полезные коктейли небольшого размера  
1.2 Предлагать в подарок игрушку, какие то комбо итд.

Продуктовые решения для второй задачи:  
2.1 Сделать большие и более питательные коктейли  
2.2 Приготавливать их заранее, реализовать варианты оплат и получения вроде "мак авто"  

"Работа" в фреймворке = задача, которую пользователь хочет решит в конкретных обстоятельствах - в контексте.  
Формулировка "человек хочет молочный коктейль" абстрактная и несет мало практической пользы.  
"Человек хочет подкрепиться пока он едет на работу" - корректное описание работы здесь.  

- Если ваше описание работы подразумевает лишь один тип продукта, который может ее решить, вероятно  
надо подняться на уровень выше в определении работы.  
- Формулировка работы не должна включать описание одного из способов ее решения (нужно это отделять)  
- Также нет смысла в слишком абстрактных дефинициях вроде высокоуровневой потребности или тренда  

https://hbr.org/2005/12/marketing-malpractice-the-cause-and-the-cure  
Каждая работа имеет социальное, эмоциональное и функциональное измерение

<u> Stakeholders chain </u>    
Когда за конечным решение стоит несколько лиц, важно понимать что их "работы" могут отличаться.  
К примеру при решении о покупке вакансии иногда принимают участие: руководитель компании (согл. бюджета),  
главный hr менеджер компании, сотрудник обладающий конечными квотами на найм.  

<u> Инерция </u>  
Чтобы клиенты поменяли способ решения своей задачи на альтернативный - тот должен быть в несколько  
раз эффективнее чем старый. Ключевой вызов для продукта:  
"стать лучше для определенных людей в определенных обстоятельствах (job to be done)". 

<u> Pain points trade-offs </u>. 

Ex: pet owner want to serve to cats taste wet food (emotional), but this food is not comfortable to prepare as a dry food (functional).
Solution: dry feed with wet center

Obstacles

- Inertia (already have solution)
- Adoption (lack of knowledge)
- Behaviour change
- Cost & Risk thresholds
- There is no market for idea

Roadmap solution checks

- How could this solution helps to fix job
- Does your solution fit with existing behaviour
- Create new or resolve old pain points
- Obstacles for this solution
- Monetisation and competition

#### Глубинные интервью

- Runnig Lean, Runnig UX (O’Relly)

Цель: получить максимально подробную картину задач, которые решают пользователи продукта  
На этом этапе важно собрать качественный сет гипотез, которые потом можно проверять количественно.  

Некоторые полезные моменты при проведении интервью:
1. наиболее ценны факты (кто что делал), а не мнения  
2. вопрос "что бы вы хотели в продукте" дает сильный bias.  
Правильнее узнавать об альтернативных методах решения проблемы и о проблеме.  
(как вы решаете эту проблему сейчас, какие способы вы для этого пробовали)
3. "dont tip your hand" не наводите клиента на ответ, формулировки должны быть открытыми
Пример: в одной фокус группе презентовали черную коробку со старой моделью техники, и новую желтую с  
прогрессивным дизайном. Клиенты в опроснике лучше оценивали новый дизайн, но когда им на выбор предложили  
забрать домой одну из коробок - большая часть выбрала черную.


<u> Кол-во интервью </u>  
Полезная эвристика:  
5 клиентов из репрезентативной рандомной выборки дадут понимание 85% имеющихся проблем в продукте  
Интервью можно повторять до тех пор, пока инсайты не начнут повторяться.
