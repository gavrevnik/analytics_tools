#### Морейнис - логика продукта (лекция)

https://www.youtube.com/watch?v=FDhgz7oBBJQ

---

1. Новое похоже на старое и должно иметь со старым ряд значительных сходств.  
Также любое новое удовлетворяет старый спрос, который в корне не меняется
2. Люди покупают то, на что уже тратят деньги и время. Основная задача - продавать новое под видом старого (например, карманный компьютер под видом телефона, потому что люди готовы покупать телефон, но у них нет статьи расходов - карманный компьютер)
3. Конкуренты есть всегда - так как у людей ограниченный запас времени и денег. Бывают косвенные и прямые конкуренты.
4. Часто покупают не продукт, а его предложение (не имея возможности с ним ознакомиться до этого). Поэтому при ресерче можно сначала продать продукт (MVP), а затем полноценно создать его
5. product/market fit в условиях текущей конкуренции за клиентов заменяется на market/product fill. Смысл в том, что в условиях конкуренции маржинальность продаж низкая, для заработка нужно масштабирование числа продаж, а значит необходимо сначала убедиться что существует для данного продукта рынок, позволяющий масштабироваться

#### Отделение Facebook Messenger в отдельное приложение

Источник: gopractise2

Интуитивно кажется, что это ухудшает пользовательский опыт, так как  
проще использовать чат в глобальном популярном приложении Facebook.  
Однако на фоне возрастающей конкуренции Whatsapp в сегменте мессенджеров,   
Facebook пришли к выводу - что отдельное приложение - это возможность эффективнее  
решить задачу клиента. Основные проблемы чата внутри общего приложения:
1. процент вкл пуш-уведомлений около 50% из за малой важности нотификаций.   
У whatsapp этот процент близок к 100, так как это важно для получения ответов  
собеседников. Отдельный мессенджер - возможность настроить такую политику по умолчанию
2. многих контактов (например родителей итд) нет в глобальном Facebook, так как они  
не используют эту соц сеть. Не хватало закрытых профилей. 

Чтобы зафорсить переток большого кол-ва людей в месседжер для того, чтобы он начал давать  
ценность - Facebook ограничил чат в основном приложении и вел на скачивание отдельного.  
Также он использовал конкурентное преимущество в виде возможности импорта графа друзей,  
а также дистрибуцию из своих основных каналов продвижения.  

#### Противостояние Teams и Слак 

Источник: gopractise2
    
Бурное развитие Слак могло оказывать отрицательный эффект на бизнес Microsoft:  
ранее клиенты пользовались корпоративной почтой, но начали переходить в рабочие чаты;  
аналогично канибализируется общение через Skype for Business  
Все это может приводить к отказу от подписок компаний на Microsoft Office 360  
(так как вкупе с google docs можно отказаться от др продуктов типа Excel итд)  

Поэтому Mirosoft запустил свой аналог Teams, много в него инвестировал и в том числе  
через свои каналы дистрибуции внедрив этот мессенджер в систему 360 успешно обогнал  
Slack и предотвратил возможные риски для своего бизнеса.  

#### Цепочки ценностей и масштабирование

Источник: gopractise1

Чтобы стать большим - надо начать с малого (малый рынок -> product fit -> масштабирование)  
Если уже найден market fit на сегменте, то наименьшие усилия - улучшать продукт так,
чтобы расширять этот сегмент на близкие сегменты.  
Анализировать паттерны поведения успешных пользователей, чтобы найти как распространить их на потенциально успешных клиентов.  
Facebook начинал как продукт для небольшой университетской среды - там он оттачивался, а после  
того как был обнаружен product-market fit, произошла экспансия вовне на более широкую ЦА.


Компании часто расширяются на рынке путем освоения дополнительных звеньев цепочки, создающей ценность потребителю. Например, Amazon в какой то момент самостоятельно стал разрабатывать свою сеть доставок.  
Поисковики разрабатывают свои браузеры, картографические сервисы и пр., замещая для пользователя своим влиянием звенья до получения желаемого результата.
"Под цепочкой создания ценности мы будем понимать структуру взаимодействия разных участников рынка, которые создают ценность, необходимую для удовлетворения конечного спроса". При этом поисковики в принципе являются центральными структурами, стоящими на пересечении большого кол-ва цепочек ценностей

#### Опыт продуктовой культуры facebook

https://gopractice.ru/skills/fb-product-culture/
Три этапа в продакт-менеджменте Facebook

1. Understand   
Глубокое изучение пользователя, продукта и рынка, которое позволит увидеть имеющиеся возможности и проблемы.  
Опирается на понимание ценности продукта и возможные запросы его клиентов.  
Анализа и изучения пользователей должно быть ровно столько, сколько нужно для принятия необходимых решений.     
Важный момент — не подгонять данные и исследования под заранее придуманные идеи,  
а искать идеи и решения исходя из полученных инсайтов о пользователе и продукте
2. Identify  
Генерация идей о том, как можно достичь цель, оценка их потенциала и приоритизация.
3. Execute  
Для наиболее приоритетных изменений начинается процесс дизайна-разработки-тестирования-релиза-анализа.
