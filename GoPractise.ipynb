### Полезные мысли 

- **Масштабирование**. Чтобы стать большим - надо начать с малого (с небольшого рынка -> fit -> масштабирование)  
Если уже найден market fit на сегменте, то наименьшие усилия - улучшать продукт так,
чтобы расширять именно этот сегмент.  
Анализировать паттерны поведения успешных пользователей, чтобы найти как распространить
их на потенциально успешных



- **Пички на графиках**. Важно обращать внимание на любые флуктуации и аномалии, возникающие в метриках на дашбордах - возможно они дадут ключ к тому как стимулировать рост метрик (пички в метриках)


- **Цепочка ценности**. Компании часто расширяются на рынке путем освоения дополнительных звеньев цепочки, создающей ценность потребителю. Например, Amazon в какой то момент самостоятельно стала разрабатывать свою сеть доставок. Поисковики разрабатывают свои браузеры, картографические сервисы и пр., замещая для пользователя своим влиянием звенья до получения желаемого результата.
"Под цепочкой создания ценности мы будем понимать структуру взаимодействия разных участников рынка, которые создают ценность, необходимую для удовлетворения конечного спроса". При этом поисковики в принципе являются центральными структурами, стоящими на пересечении большого кол-ва цепочек ценностей



### Product Market Fit & Retention

Ключевой шаг перед масштабированием продукта - понять достиг ли он market fit.  

**Market fit** = ваш продукт лучше решает проблемы людей по сравнению с прежними решениями.  
<u>Критерии достижения product market fit </u>:

- Retention выходит на плато (часть привлеченных клиентов стабильно возвращается)  
Важна гранулярность (дни/недели/..)  - исходя из частотности проблемы людей которая решается. 


- Если Retention не выходит на плато -> его затухание сравнимо/лучше чем у успешных конкурентов в отрасли.  
Есть бизнесы в которых невозможно достижение плато (дейтинговые сервисы, покупки квартиры итд)

Два типа бизнесов:  

- регулярные задачи (выход в почту, соц сети итд)    
Сложнее выиграть у старой привычки, так как есть привыкание  
Надо быть сильно лучше чем на рынке
Зато проще удерживать засчет постоянной зависимости


- нерегулярные задачи (покупка квартиры)  
Легче выиграть так как пользователь выбирает в моменте.  
Сложнее удержать, так как в след. раз он опять будет выбирать заново  
<u> Решение:</u> привлечь клиента в смежные сервисы с большей частотностью задач  
(пример - сервис по отслеживанию цен на жилье VS сервис по покупке жилья)  
Перманентная реклама чтобы быть top of mind (например, aviasales)


Один продукт может иметь ретеншн на одном сегменте (страны / клиенты итд) и не иметь на другом.  
Самый простой способ - это растить потенциальный рынок, на котором ты уже имеешь market fit  
(!) Retention - это функция от продукта и от пользователей! Нерелевантные пользователи с хорошим  
продуктом также не образуют устойчивый product market fit

Еще один подход к <u> измерению </u> product-market fit: https://gopractice.ru/stories/superhuman_engine/  
Опрос: "как люди будут себя чувствовать если их лишить данного продукта?"  
Если разочарованных >40%, то с большой вероятностью этот продукт имеет market fit в этой ЦА.  

Далее возможен следующий алгоритм для расширения сегмента с достижением market_fit:  
- понять что за клиенты - те, кто разочаровались бы без этого продукта (что их описывает)  
- за что они любят продукт? что мешает близким к ним клиентам любить этот продукт?  

Далее роадмап улучшений строится над идеями улучшать то что любят, устранять из за чего не любят

**Market fit - непрерывная функция**  
Чем сильнее market fit - тем чаще пользователь прибегает к продукту для решения своей задачи.  
При этом макс. частота прибегания ограничена частотой самой задачи. А это определяет Retention.  
К примеру дневной Retention для людей, покупающих продукты в среднем раз в два дня, будет ~50%  
Если сравнить два продукта <u> на одинаковой аудитории </u> решающих одинаковые задачи и  
Retention одного больше другого - то это означает, что его product-market fit сильнее.
